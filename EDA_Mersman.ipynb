```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#load dataframe and view first 5 rows
df= pd.read_csv('heart_data.csv', index_col = 0)
df.head()
```

```python
#view last 5 rows
df.tail()
```

```python
#convert age from days to years
df.age = [x/365 for x in df.age]
df.age = df.age.astype(int)
```

```python
#review dataframe
df.head()
```

```python
#convert df.gender from numerical to categorical, Male - 1, Female - 2 
df.gender = np.where(df.gender == 1, 'Male', 'Female')
```

```python
#confirm 
df.head()
```

```python
#create BMI column
df['BMI'] = df['weight']/((df['height']/100)**2)
df['BMI'] = df['BMI'].astype(int)
```

```python
#use .describe() to view descriptive statistics for numerical variables
df.describe()
```

The highest BMI recorded was 105. 

```python
df = df[df.BMI < 105]
```

```python
df[df.BMI >55]
```

```python
#review ap_hi for anything under 0 since the patient would be dead
df[df.ap_hi <= 0]
```

```python
# change sign, these appear to be simple sign errors based on the corresponding ap_lo
df.ap_hi = np.where(df.ap_hi < 0, abs(df.ap_hi), df.ap_hi)
df.ap_hi.describe()
```

```python
#review ap_lo for anything under 0 
df[df.ap_lo <= 0]
```

```python
#remove patients whose ap_lo was 0 or below, this data does not make sense
df = df[df.ap_lo > 0]
```

## According to pubmed -- "The highest pressure recorded in an individual was 370/360." https://pubmed.ncbi.nlm.nih.gov/7741618/
It is assumed anything over these pressure is an entry error and will be excluded.

```python
#Subset the Dataframe for only possible bp values
df = df[df.ap_hi <= 370]
df = df[df.ap_lo <= 360]
df.describe()
```

```python
#Trends for age:
plt.figure(figsize=(10,5))
sns.countplot(x=df['age'],hue='cardio', data=df, alpha=0.75)
plt.show()
```

Over the age of 55 cardiovascular disease becomes more prevalent.

```python
#Trends for age and cholesterol
plt.figure(figsize=(10,5))
sns.lineplot(x=df['age'], y=df['cholesterol'], hue = 'cardio',  data=df, alpha=0.75)
plt.show()
```

People with cardiovascular disease are more apt to have higher cholesterol, as people get older their cholesterol typically increases but the difference between cholesterol levels of healthy vs cardiovascular diseased are still apparent.

```python
#Trends for age and BMI:
plt.figure(figsize=(10,5))
sns.lineplot(x=df['age'], y=df['BMI'], hue = 'cardio',  data=df, alpha=0.75)
plt.show()
```

People with cardiovascular disease are more apt to have higher BMI, as people get older their BMI typically increases but the difference between BMI status of healthy vs cardiovascular diseased are still apparent.

```python
#Trends for cholesterol and BMI:
plt.figure(figsize=(10,5))
sns.lineplot(x=df['cholesterol'], y=df['BMI'], data=df, alpha=0.75)
plt.show()
```

Cholesterol increases with BMI but there are changes in that relationship from cholesterol category 1-2 and 2-3. 

```python
#Trends for age:
plt.figure(figsize=(10,5))
sns.lineplot(x=df['cardio'], y=df['cholesterol'], data=df, alpha=0.75)
plt.show()
```

```python
#Trends for age:
plt.figure(figsize=(10,5))
sns.lineplot(x=df['cardio'], y=df['BMI'], data=df, alpha=0.75)
plt.show()
```

An increased BMI increases the likelihood of cardiovasular disease.

```python
#split dataframes into healthy and cardiovascular disease to compare histograms
heart_disease = df[df.cardio == 1]
healthy = df[df.cardio == 0]
```

## View the dispersion comparisons for cardiovascular disease vs non-cardiovascular disease

```python
ax = healthy.age.hist(alpha=0.5)
heart_disease.age.hist(ax=ax, alpha = 0.75)
```

```python
#define function to plot the columns
def plot_comparison(column):
    ax = healthy[column].hist(alpha=0.75)
    heart_disease[column].hist(ax=ax, alpha = 0.75)
    plt.title("Health vs Heart Disease Comparing: " + column)
    ax.legend(['Healthy', 'CV Disease'])
    plt.show()

    
    
```
